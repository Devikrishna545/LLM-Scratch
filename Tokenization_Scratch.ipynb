>Tokenization



```python
#first loading the data
data = open("the-verdict.txt","r",encoding="utf-8")
raw_text = data.read()
print(raw_text[:99])
```

```python
#find the length of raw_text
print(len(raw_text))
```

```python
#split the text with white spaces using "re" regex expression library
import re
preprocessed = re.split(r'(\s)',raw_text)
print(preprocessed[:99])
```

```python
#should split with commas and periods and whitespaces and additional special characters
preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)',raw_text)
print(preprocessed[:99])
```

```python
#In our preprocesssing we are removing these white spaces as it is not required
preprocessed = [item.strip()for item in preprocessed if item.strip()]
print(preprocessed[:99])
```

```python
#length of preprocessed
len(preprocessed)
```

>Token ID's

```python
#Now adding id's to token in the sorted manner, making these to a set so no duplication of ids happen.
#all words in set
all_words = sorted(set(preprocessed))
len(all_words)
```

```python
#adding ID's to this all_words
vocab = {token:integer for integer, token in enumerate(all_words)}
```

```python
#printing the vocab
for i,item in enumerate(vocab.items()):
  print(item)
  if i>50:
    break
```

```python
#we need to create a function to encode and decode the words of a sentence.
class SimpleTokenizerV1():
  def __init__(self,vocab):
    self.str_to_int = vocab
    self.int_to_str = {integer:token for token,integer in vocab.items()}

  def encode(self,text):
    preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)',text)
    preprocessed = [item.strip()for item in preprocessed if item.strip()]
    Ids =[self.str_to_int[item] for item in preprocessed]
    return Ids

  def decode(self,Ids):
    text = " ".join([self.int_to_str[Id] for Id in Ids])
    #replace space before specified punctuations
    text = re.sub(r'\s+([,.?!"()\'])',r'\1',text)
    return text
```

```python
#testing tokenizerV1
text = """"It's the last he painted, you know,"
           Mrs. Gisburn said with pardonable pride."""
tokenizer = SimpleTokenizerV1(vocab)
Ids = tokenizer.encode(text)
print(Ids)
```

```python
raw_data = tokenizer.decode(Ids)
print(raw_data)
```

```python
#adding special context tokens
#we have to add tokens for unknown and end of sequence
#adding tokens to vocab
all_words.extend(["<|endoftext|>", "<|unk|>"])
vocab = {token:integer for integer, token in enumerate(all_words)}
len(vocab)
```

```python
#will create another class SimpleTokenizerV2

class SimpleTokenizerV2():
  def __init__(self,vocab):
    self.str_to_int = vocab
    self.int_to_str = {integer:token for token,integer in vocab.items()}

  def encode(self,text):
    preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)',text)
    preprocessed = [item.strip()for item in preprocessed if item.strip()]
    preprocessed = [item if item in self.str_to_int
                    else "<|unk|>" for item in preprocessed]
    Ids = [self.str_to_int[item] for item in preprocessed]
    return Ids

  def decode(self,Ids):
    text = " ".join([self.int_to_str[Id] for Id in Ids])
    #replace space before specified punctuations
    text = re.sub(r'\s+([,.?!"()\'])',r'\1',text)
    return text
```

```python
#testing same sample
tokenizer = SimpleTokenizerV2(vocab)

sample = "Hello world, How are you?"
Ids = tokenizer.encode(sample)
print(Ids)
```

```python
#decoding the above sample
raw_data = tokenizer.decode(Ids)
print(raw_data)
```

## **Datasets** and **Dataloader**

```python
#install pytorch
# to get the input and output tensors , we use pytorch libraries dataset and dataloader.
from torch.utils.data import Dataset, DataLoader

#steps
#1.tokenize the complete texts
#2.chunking of the text

class GPTDatasetV1(Dataset):
  def __init__(self,txt,tokenizer,max_length,stride):
    self.input_ids =[]
    self.output_ids=[]

    #tokenizer
    token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})

    # using a sliding window to make the input and output chunks
    # Use a sliding window to chunk the book into overlapping sequences of max_length
    for i in range(0,len(token_ids)-max_length,stride):
      input_chunks = token_ids[i:i+max_length]
      output_chunks = token_ids[i+1: i+ max_length +1]
      self.input_ids.append(torch.tensor(input_chunks))
      self.output_ids.append(torch.tensor(output_chunks))

  def __len__(self):
    return len(self.input_ids)

  def __getitem__(self,idx):
    return self.input_ids[idx],self.output_ids[idx]
```

```python
def create_dataloader_v1(txt, batch_size=4, max_length=256,
                         stride=128, shuffle=True, drop_last=True,
                         num_workers=0):
  #initialize tokenizer
  tokenizer = tiktoken.get_encoding("gpt2")
  #create dataset
  dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)
  #create data loader
  dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle,
                          drop_last=drop_last,num_workers=num_workers)
  return dataloader
```

```python
#open file to get raw text
data = open("the-verdict.txt","r",encoding="utf-8")
raw_text = data.read()
```

```python
import torch
#data loader for batch size 8
dataloader = create_dataloader_v1(raw_text,batch_size=8,max_length=4,stride=4,shuffle=False)

new_iter = iter(dataloader)
inputs, targets = next(new_iter)
print("Inputs:\n", inputs)
print("\nTargets:\n", targets)
```
